```python
import scipy as sp
import matplotlib.pyplot as plt
```

```python
A_r = 100.
phase = sp.linspace(0, 2*sp.pi,500)  # 0.71399
```

```python
c = sp.sqrt(A_r)*sp.cos(phase)
s = sp.sqrt(A_r)*sp.sin(phase)
```

```python
%matplotlib inline
plt.figure(figsize=(10,10))
ax1 = plt.subplot(221)
ax1.plot(phase, c)

ax2 = plt.subplot(222)
ax2.plot(phase, s)
```
